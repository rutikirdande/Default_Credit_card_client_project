```python
import pandas as pd
```

```python
df = pd.read_csv('./data/default of credit card clients.csv', skiprows=1, header=None)

# set the first row as the header
df.columns = df.iloc[0]
df = df[1:]

# display the resulting dataframe
df.head()
```

```python
df=df.drop(labels=['ID','SEX','EDUCATION','MARRIAGE'], axis=1)
```

```python
## check for duplicated records
df.duplicated().sum()

# Extract duplicate rows
df.loc[df.duplicated(), :]

df.drop_duplicates()
```

```python
## Independent and dependent features
x = df.drop(labels=['default payment next month'],axis=1)
y = df[['default payment next month']]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
x,y=make_classification(n_samples=1000,n_features=5,random_state=123)
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model=LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
